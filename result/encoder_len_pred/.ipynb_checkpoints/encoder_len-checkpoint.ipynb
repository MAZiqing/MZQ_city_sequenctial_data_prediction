```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df_ori = pd.read_csv('./encoder_len.csv', encoding='gbk')
df_ori
```

```python
df_ori.pivot(values='best_valid_')
```

```python
df = df_ori.copy()
plt.figure(figsize=(6,4))
for i in range(0,4):
    list_all = df.loc[i, 'recon_error'][1:-1].split(',')
    list_mse = [float(i[7:18]) for i in list_all[1:9:2]]
    if i == 3:
        list_mse = list_mse[1:]
    plt.plot(list_mse,'.-')
plt.legend(list(df['model_name']))
plt.legend(['DS-RNN', 'DS-RNN-II', 'DSTP-RNN', 'dSTA-RNN'])
plt.xticks([0,1,2,3])
plt.xlabel('decoder step')
plt.ylabel('MSE Loss')
# plt.xticklabels([1,2,3,4])

# plt.ylim(0, 0.01)
# plt.savefig('./step_pred.eps', format='eps')
plt.show()
```
