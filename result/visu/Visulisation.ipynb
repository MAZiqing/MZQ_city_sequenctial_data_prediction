```python
def plot_att(df):
    fig, ax = plt.subplots(1,1, figsize=(4,8))
    fig.set_facecolor('w')
#     fig.set_figwidth(40)
#     fig.set_figheight(80)
    # ax.set_facecolor('w')
    df_index = ['F'+str(i) for i in range(1,12)] + ['P'+str(j) for j in range(1,8)]
    df.index = df_index
    img = ax.imshow(df.values, cmap='Blues')
    ax.set_yticks(list(range(0, df.shape[0])))
    ax.set_yticklabels(list(df.index))
    ax.set_aspect(7)
    ax.set_xlabel('Encoder step')
    ax.set_ylabel('Input series')
#     plt.colorbar(img)
    plt.savefig('./sa_score.png', format='png',dpi=500)
    plt.show()
plot_att(dfs)
# plot_att(dft)

# fig, ax = plt.subplots(1,1)
# ax.imshow(dft.values)
# # ax.set_yticks(list(range(0, 18)))
# # ax.set_yticklabels(list(df.index))
# plt.show()
```
