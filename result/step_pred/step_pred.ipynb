```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df_ori = pd.read_csv('./best.csv', encoding='gbk')
df_ori
```

```python
df = df_ori.copy()
plt.figure(figsize=(4,3))
for i in [0,2,1,3]:
    list_all = df.loc[i, 'recon_error'][1:-1].split(',')
    list_mse = [float(i[7:18]) for i in list_all[1:23:2]]
#     if i == 3:
#         list_mse = list_mse[1:]
    plt.plot(list_mse,'.-')
# plt.legend(list(df['model_name']))
plt.legend(['DSTP-RNN', 'DS-RNN', 'DS-RNN-II', 'hDS-RNN'])
# plt.xticks([0,1,2,3])
plt.xlabel('Decoder step')
plt.ylabel('MSE Loss')
plt.tight_layout()
# plt.xticklabels([1,2,3,4])

# plt.ylim(0, 0.01)
plt.savefig('./step_pred.eps', format='eps')
plt.show()
```

```python
df = df_ori.copy()
plt.figure(figsize=(4,3))
for i in [0,2,1,3]:
    list_all = df.loc[i, 'recon_error'][1:-1].split(',')
    list_mae = [float(i[9:18]) for i in list_all[0:22:2]]
#     if i == 3:
#         list_mse = list_mse[1:]
    plt.plot(list_mae,'.-')
# plt.legend(list(df['model_name']))
plt.legend(['DSTP-RNN', 'DS-RNN', 'DS-RNN-II', 'hDS-RNN'])
# plt.xticks([0,1,2,3])
plt.xlabel('Decoder step')
plt.ylabel('MAE Loss')
# plt.xticklabels([1,2,3,4])
plt.tight_layout()
# plt.ylim(0, 0.01)
plt.savefig('./step_pred_mae.eps', format='eps')
plt.show()
```
