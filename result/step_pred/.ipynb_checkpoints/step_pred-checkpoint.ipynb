```python
plt.figure()
for i in range(0,4):
    list_all = df.loc[i, 'recon_error'][1:-1].split(',')
    list_mse = [float(i[7:18]) for i in list_all[1:9:2]]
    plt.plot(list_mse)
plt.legend(list(df['model_name']))
plt.ylim(0, 0.01)
plt.show()
#     plt.savefig('./step_pred.eps', format='eps')
```
